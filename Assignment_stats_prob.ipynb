```python
#STATISTICS:

import numpy as np
import pandas as pd
# Q.1:
#given sample :
#3, 2, 5, 4, 7, 2, 3, 3, 1, 6, 4, 2, 3, 5, 2, 4, 2, 1, 3, 5, 6, 3, 2, 1,
# 4, 2, 4, 5, 3, 2, 7, 2, 3, 4, 5, 1, 6, 2, 4, 3, 5, 3, 2, 4, 2, 6, 3, 2, 4, 5

#a. Mean: What is the average rental duration for customers at the car rental company?
data = [3, 2, 5, 4, 7, 2, 3, 3, 1, 6, 4, 2, 3, 5, 2, 4, 2, 1, 3, 5, 6, 3, 2, 1,
        4, 2, 4, 5, 3, 2, 7, 2, 3, 4, 5, 1, 6, 2, 4, 3, 5, 3, 2, 4, 2, 6,
        3, 2, 4, 5]
mean_duration = np.mean(data)
print("Mean rental duration is:\n", mean_duration)

#b.Median: What is the typical or central rental duration experienced by customers?
median_duration = np.median(data)
print("Median rental duration is:\n", median_duration)

#c.Mode: Are there any recurring or most frequently occurring rental durations for customers?
series_data = pd.Series(data)
mode_duration = series_data.mode()[0]
print("Mode rental duration is:\n", mode_duration)
```
